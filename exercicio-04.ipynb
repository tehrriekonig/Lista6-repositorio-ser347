```python
import matplotlib.pyplot as plt
```

```python
idades = ["0 a 4 anos", "5 a 9 anos", "10 a 14 anos ", "15 a 19 anos ", "20 a 24 anos ", "25 a 29 anos ", "30 a 34 anos ", "35 a 39 anos ", "40 a 44 anos ", "45 a 49 anos ", "50 a 54 anos ", "55 a 59 anos ", "60 a 64 anos ", "65 a 69 anos ", "70 a 74 anos ", "75 a 79 anos ", "80 a 84 anos ", "85 a 89 anos ", "90 a 94 anos ", "95 a 99 anos ", "100 anos e mais"]

masculino = [7016987, 7624144, 8725413, 8558868, 8630229, 8460995, 7717658, 6766664, 6320568, 5692014, 4834995, 3902344, 3041035, 2224065, 1667372, 1090517, 668623, 310759, 114964, 31529, 7247]

feminino = [6779171, 7345231, 8441348, 8432004, 8614963, 8643419, 8026854, 7121915, 6688796, 6141338, 5305407, 4373877, 3468085, 2616745, 2074264, 1472930, 998349, 508724, 211594, 66806, 16989]

```

```python
x_pos = [ x for x in range( len(idades) ) ]

plt.figure(figsize=(10,8))

fig, axes = plt.subplots(ncols=2, sharey=True)

axes[0].barh(x_pos, masculino, align='center', color='Blue')
axes[1].barh(x_pos, feminino, align='center', color='Red')
axes[0].invert_xaxis()
axes[0].yaxis.tick_right()

plt.setp(axes, xticks=[ 0, 3000000, 6000000, 9000000], xticklabels=[ "0", "3 mi", "6 mi", "9 mi"],
        yticks=x_pos, yticklabels=idades)


# Ajuste de posição do título
st = fig.suptitle("Distribuição das Populações Femininas e Masculinas do Brasil, 2010")
st.set_y(1.1)
st.set_x(0.2)

# Remoção das linhas pretas em volta de cada subplot
axes[0].spines["top"].set_visible(False)
axes[0].spines["left"].set_visible(False)
axes[0].spines["bottom"].set_visible(False)

axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)
axes[1].spines["bottom"].set_visible(False)

fig.tight_layout()
fig.subplots_adjust(left=-0.6, wspace=0.32)
```
